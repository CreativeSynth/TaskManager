```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
from time import sleep
import math

import warnings
warnings.filterwarnings(action='ignore')
```

```python
# get korean_quiz from sixth_korean_quiz.csv
df = pd.read_csv('sixth_korean_quiz.csv')
korean_quiz = df
korean_quiz
```

```python
# driver option 설정
options = webdriver.ChromeOptions()
options.headless = False # 크롬창을 띄우지 않고 크롤링 하려면 이 값을 True로 바꾸기
options.add_argument("window-size=1920x1080")
# ChromeDriver 열기
web_url = "https://wquiz.dict.naver.com/quiz.dict?service=krdic&dictType=koko&quizId=14d22301d66544c899a24934aa758982&group_id=1&sort_type=1"

chrome_driver_path = "./chromedriver"
chrome_service = webdriver.chrome.service.Service(chrome_driver_path)
browser = webdriver.Chrome(service=chrome_service, options=options)

# open url
browser.get(web_url)
```

```python
quiz_id = web_url.split('quizId=')[1].split('&')[0]
korean_quiz.loc[0,'id'] = quiz_id

category_elem = browser.find_element(By.XPATH, r'//*[@id="quiz_main"]/h2/span')
question_elem = browser.find_element(By.XPATH, r'//*[@id="vote_name_id"]')
korean_quiz.loc[0,'category'] = category_elem.text
korean_quiz.loc[0,'question'] = question_elem.text
```

```python
browser.find_element(By.XPATH, r'//*[@id="answer_select_area"]/a[1]').click()
```

```python
crct_answer = browser.find_element(By.CLASS_NAME, r'crct').text
crct_answer = crct_answer.replace('%','').replace('\n','')
crct_answer = ''.join([i for i in crct_answer if not i.isdigit()])
korean_quiz.loc[0,'answer'] = crct_answer
korean_quiz.loc[0,'select1'] = crct_answer

wrong_answers = browser.find_elements(By.CLASS_NAME, r'wrong')
wrong_answers = [i.text.replace('%','').replace('\n','') for i in wrong_answers]
wrong_answers = [''.join([i for i in j if not i.isdigit()]) for j in wrong_answers]
for i in range(0, len(wrong_answers)):
    korean_quiz.loc[0,'select'+str(i+2)] = wrong_answers[i]

korean_quiz
```

```python
# find next button and click
next_button = browser.find_element(By.CLASS_NAME, r'btn_next')
next_button.click()
```

```python
i = 3145
while (i < 3133 + 52):
    existing_quiz = False
    # wait for page load
    WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.CLASS_NAME, r'quiz_tit')))
    
    # get quiz id
    quiz_id = browser.current_url.split('quizId=')[1].split('&')[0]

    # if a quiz with the same id is in korean_quiz, 
    if quiz_id in korean_quiz['id'].values:
        existing_quiz = True
    else:
        i += 1
        korean_quiz.loc[i,'id'] = quiz_id
        # get question and category
        category_elem = browser.find_element(By.XPATH, r'//*[@id="quiz_main"]/h2/span')
        question_elem = browser.find_element(By.XPATH, r'//*[@id="vote_name_id"]')
        korean_quiz.loc[i,'category'] = category_elem.text
        korean_quiz.loc[i,'question'] = question_elem.text
        korean_quiz['question'] = korean_quiz['question'].str.replace('\n','')
        
    # next page
    browser.find_element(By.XPATH, r'//*[@id="answer_select_area"]/a[1]').click()
    
    # wait for page load
    WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.CLASS_NAME, r'crct')))

    if existing_quiz is False: 
    # get answer and all selections
        crct_answer = browser.find_element(By.CLASS_NAME, r'crct').text
        crct_answer = crct_answer.replace('%','').replace('\n','')
        crct_answer = ''.join([j for j in crct_answer if not j.isdigit()])
        korean_quiz.loc[i,'answer'] = crct_answer
        korean_quiz.loc[i,'select1'] = crct_answer

        wrong_answers = browser.find_elements(By.CLASS_NAME, r'wrong')
        wrong_answers = [k.text.replace('%','').replace('\n','') for k in wrong_answers]
        wrong_answers = [''.join([k for k in j if not k.isdigit()]) for j in wrong_answers]
        for j in range(0, len(wrong_answers)):
            korean_quiz.loc[i,'select'+str(j+2)] = wrong_answers[j]
    
    existing_quiz = False
    
    # wait for page load
    WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.CLASS_NAME, r'btn_next')))
    # move to next quiz
    next_button = browser.find_element(By.CLASS_NAME, r'btn_next')
    next_button.click()

korean_quiz.to_csv('n_korean_quiz.csv', index=False, encoding='utf-8-sig')
```

```python
# close chrome driver
browser.quit()
```

```python
# from korean_quiz.cs get korean_quiz
df = pd.read_csv('korean_quiz.csv')
```

```python
# ko_quiz_1(알쏭달쏭 우리말)
ko_quiz_1 = pd.DataFrame(columns=['task_name', 'index', 'prompt', 'answer', 'attribute'])

i = 0 # index
while (i < 1516):
    # get index 
    ko_quiz_1.loc[i,'index'] = i
    # get prompt
    ko_quiz_1.loc[i,'prompt'] = df.loc[i,'question']
    # get answer
    ko_quiz_1.loc[i,'answer'] = df.loc[i,'answer']
    # append to ko_quiz_1
    ko_quiz_1.loc[i,'task_name'] = 'ko_quiz_1'
    i += 1
ko_quiz_1
```

```python
# ko_quiz_2(올바른 띄어쓰기)
ko_quiz_2 = pd.DataFrame(columns=['task_name', 'index', 'prompt', 'answer', 'attribute'])

i = 0 # index
df_index = 1516
while (df_index < 1701):
    ko_quiz_2.loc[i,'index'] = i
    # from column 'select1' to 'select3', concat to prompt
    for j in range(1,4):
        if pd.isnull(df.loc[df_index,'select'+str(j)]) is False:
                ko_quiz_2.loc[i,'prompt'] = str(ko_quiz_2.loc[i,'prompt']) + '"' + str(df.loc[df_index,'select'+str(j)]) + '" '
    ko_quiz_2.loc[i,'prompt'] = ko_quiz_2.loc[i,'prompt'].replace('nan','')
    ko_quiz_2.loc[i,'prompt'] = ko_quiz_2.loc[i,'prompt'] + '중 ' + df.loc[df_index,'question']
    ko_quiz_2.loc[i,'answer'] = df.loc[df_index,'answer']
    ko_quiz_2.loc[i,'task_name'] = 'ko_quiz_2'
    i += 1
    df_index += 1

ko_quiz_2
```

```python
# ko_quiz_3(알듯말듯 우리속담)
ko_quiz_3 = pd.DataFrame(columns=['task_name', 'index', 'prompt', 'answer', 'attribute'])

i = 0 # index
df_index = 1701
while (df_index < 1823):
    ko_quiz_3.loc[i,'index'] = i
    # from column 'select1' to 'select3', concat to prompt
    if (type(df.loc[df_index,'select2']) == str) and (pd.isna(df.loc[df_index,'select2']) is True):
            ko_quiz_3.loc[i,'prompt'] = df.loc[df_index,'question'] + '에서 ? 에 들어갈 단어로만 대답하시오.'     
    elif (type(df.loc[df_index,'select2'])== float) and (math.isnan(df.loc[df_index,'select2']) is True):
            ko_quiz_3.loc[i,'prompt'] = df.loc[df_index,'question'] + '에서 ? 에 들어갈 단어로만 대답하시오.'
    else:
        for j in range(1,4):
            if pd.isnull(df.loc[df_index,'select'+str(j)]) is False:
                    ko_quiz_3.loc[i,'prompt'] = str(ko_quiz_3.loc[i,'prompt']) + '"' + str(df.loc[df_index,'select'+str(j)]) + '" '
        ko_quiz_3.loc[i,'prompt'] = ko_quiz_3.loc[i,'prompt'].replace('nan','')
        ko_quiz_3.loc[i,'prompt'] = ko_quiz_3.loc[i,'prompt'] + '중 ' + df.loc[df_index,'question'] + '에서 ? 에 들어갈 단어로만 대답하시오.'
    ko_quiz_3.loc[i,'answer'] = df.loc[df_index,'answer']
    ko_quiz_3.loc[i,'task_name'] = 'ko_quiz_3'
    i += 1
    df_index += 1

ko_quiz_3
```
